### LONGITUDE VALUES ARE NEGATIVE. HENCE CONVERTED TO POSITIVE
IT DECREASES EUCLIDEAN DISTANCE IN DIMENSIONAL FEATURE SPACE W.R.T TO OTHER ATTIBUTES

```python
df['longitude']= df['longitude']* -1
```

###  CATEGORICAL DATA CONVERTED TO NUMERICAL
(CATEGORICAL DATA CANNOT BE EVALUATED DIRECTLY)

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
df.head()
```

### FINDING NUMBER OF OUTLIERS IN NUMERICAL DATA ATTRIBUTES BY IQR METHOD

```python
df1 = df[['speedlimit_mph','speed_mph','latitude','longitude','functionalClass']]
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).sum()
```

### PEARSON CORRELATION
IT IS A NUMBER BETWEEN -1 AND 1 THAT INDICATES THE EXTENT TO WHICH TWO VARIABLES ARE LINEARLY RELATED 
BELOW TABLE SHOWS THE CORRELATION OF EACH ATTRIBUTE WITH RESPECT TO EVERY OTHER ATTRIBUTE.
IF TWO VARIABLES ARE HIGHLY CORRELATED, THEN ONE OF THEM CAN BE DROPPED.

```python
df.corr(method='pearson').style.format("{:.3}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```
